# 3. Data Preprocessing

```python

# Replacing zero values with NaN
diabetes_ = df_diabetes
```

```python
diabetes_[["Glucose",
           "BloodPressure",
           "SkinThickness", 
           "Insulin", "BMI"]] = diabetes_[["Glucose", 
                                           "BloodPressure",
                                           "SkinThickness", "Insulin",
                                                               "BMI"]].replace(0, np.NaN) 
```

```python
diabetes_.head(5)
```

```python
diabetes_.tail(10)
```

```python
#Count of Nan 
df_diabetes.isnull().sum()
```

```python
# Replacing NaN with mean values
diabetes_["Glucose"].fillna(diabetes_["Glucose"].mean(), inplace = True)
diabetes_["BloodPressure"].fillna(diabetes_["BloodPressure"].mean(), inplace = True)
diabetes_["SkinThickness"].fillna(diabetes_["SkinThickness"].mean(), inplace = True)
diabetes_["Insulin"].fillna(diabetes_["Insulin"].mean(), inplace = True)
diabetes_["BMI"].fillna(diabetes_["BMI"].mean(), inplace = True)
```

```python
#Statistical summary
diabetes_.describe().T
```

-> Features : features is known as independent variable that predict the dependent variable label

-> Label :The prediction we done is known as Label 

-> Target : Target is the final output an ML model is trying to predict. 
         It can be categorical or continuous.The features describe the label, 
          so they should also describe the target

-> Difference Between Label & Feature : a label represents an output value, 
                      while a feature is an input value that describes 
                       the characteristics of such labels in datasets.
                       

        

```python
#Feature Scaling using MinMaxScaler
#Feature Scaling is the process of normalising the range of features in a dataset 

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler( feature_range=(0, 1))

diabetes_scaled = scaler.fit_transform(diabetes_)
diabetes_scaled = pd.DataFrame(diabetes_scaled)
```

```python
diabetes_scaled
```

```python
#Selecting Features - ['Glucose','Insulin','BMI','Age']

X = diabetes_scaled.iloc[:,[1,4,5,7]]
y = diabetes_scaled.iloc[:,8]  #Outcome
```

```python
X
```

```python
list(y)
```

```python
#splitting X and y
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,stratify = diabetes_['Outcome'])
```

```python
#Checking dimensions 
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# 4. ML Model 

Here i use use the scikit-learn library to create ML models. 
When coding, this library is written as sklearn.

The steps to building and using a model are:

1 - Define:  What type of model will it be? A decision tree? Some other type of model?
         #Some other parameters of the model type are specified too.
    
2 - Fit: Capture patterns from provided data. This is the heart of modeling.

3 - Predict: Just what it sounds like

4 - Evaluate: Determine how accurate the model's predictions are.



```python
# Model 1 : Logistic Regression 
#fit the data 

from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
y_pred 
```

```python
#Model 2 : Decision Tree 
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier()
dectree.fit(X_train,y_train)
```

```python
#Model 3 - Random Forest 
from sklearn.ensemble import RandomForestClassifier
ranfor = RandomForestClassifier()
ranfor.fit(X_train, y_train)
```

```python
#Model 4 - Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
#Model 5 - Support Vector Classifier
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
```

```python
#Model 6 - K- Nearest neighbour
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# Making predictions on test dataset
y_pred_reg = reg.predict(X_test)
y_pred_dectree = dectree.predict(X_test)
y_pred_svc = svc.predict(X_test)
y_pred_nb = nb.predict(X_test)
y_pred_knn = knn.predict(X_test)
y_pred_ranfor = ranfor.predict(X_test)
```

# 5. Model Evaluation

1.Model evaluation is the process that uses some metrics which
help us to analyze the performance of the model.

2.There are many metrics like Accuracy, Precision, Recall, F1 score, 
#Area under Curve, Confusion Matrix, and Mean Square Error. 

3.Cross Validation is one technique that is followed during the training phase
#and it is a model evaluation technique as well.

```python
from sklearn.metrics import accuracy_score
```

```python

accuracy_reg = accuracy_score(y_test, y_pred_reg)
accuracy_knn = accuracy_score(y_test,y_pred_svc)
accuracy_nb = accuracy_score(y_test,y_pred_nb)
accuracy_dectree = accuracy_score(y_test,y_pred_dectree)
accuracy_ranfor = accuracy_score(y_test,y_pred_ranfor)
accuracy_svc = accuracy_score(y_test,y_pred_svc)
```

Here 'y_test 'represents the 'true labels', &  'y_pred 'represents the' predicted labels'.

The 'accuracy_score' function compares these two arrays & 
returns the accuracy of the model's predictions.

```python
# Accuracy on test set

print("Logistic Regression: {:.2f}%".format(accuracy_reg * 100))
print("K-Nearest neighbors:{:.2f}%".format(accuracy_knn*100))
print("Support Vector Classifier:{:.2f}%".format(accuracy_svc*100))
print("Naive Bayes:{:.2f}%".format(accuracy_nb*100))
print("Decision tree:{:.2f}%".format(accuracy_dectree*100))
print("Random Forest:{:.2f}%".format(accuracy_ranfor*100))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_knn)
cm
```

```python
# Heatmap of Confusion matrix
sns.heatmap(pd.DataFrame(cm), annot=True)
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_knn))
```
